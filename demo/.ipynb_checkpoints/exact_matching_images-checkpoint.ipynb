```python
import os
from imutils import paths
import time
import cv2
import glob
import shutil
import os
import imagewizard as iw
```

```python
ALL_IMAGES_PATH_TO_DIR = "data\\all_images"
SAMPLE_IMAGES_PATH_TO_DIR = "data\\sample_images"
```

```python
print("[INFO] computing hashes for all images in dataset")

all_images_paths = list(paths.list_images(ALL_IMAGES_PATH_TO_DIR))
sample_images_paths = list(paths.list_images(SAMPLE_IMAGES_PATH_TO_DIR))

BASE_PATHS = set([p.split(os.path.sep)[-2] for p in sample_images_paths])
```

```python
len(all_images_paths)
```

```python
all_images_hashes = {}
start = time.time()
```

```python
hashing = iw.Hashing()
sim = iw.Similarity()
```

```python
# loop over the haystack paths
for p in all_images_paths:
    # load the image from disk
    image = cv2.imread(p)

    # if the image is None then we could not load it from disk (so
    # skip it)
    if image is None:
        continue
    
    # compute the hash
    imageHash = hashing.dhash(image=image, order='bgr')

    # update the haystack dictionary
    l = all_images_hashes.get(imageHash, [])
    l.append(p)
    all_images_hashes[imageHash] = l
```

```python
# show timing for hashing haystack images, then start computing the
# hashes for needle images
print("[INFO] processed {} images in {:.2f} seconds".format(len(all_images_hashes), time.time() - start))
print("[INFO] computing hashes for sample images...")
```

```python
# loop over the sample image paths
for p in sample_images_paths:
    # load the image from disk
    image = cv2.imread(p)

    # if the image is None then we could not load it from disk (so
    # skip it)
    if image is None:
        continue

    # compute the hash
    imageHash = hashing.dhash(image=image, order='bgr')

    for ih in all_images_hashes:
        score = sim.similarity(metric='hamming', value_query=imageHash, value_src=ih)
        
        if score <= 10:
            matched_images_path = all_images_hashes.get(ih, [])[0]

            base_dir = matched_images_path.split(os.path.sep)[-2]
            img_name = matched_images_path.split(os.path.sep)[-1]   
            dst_dir = os.path.join("data", "results", base_dir)
            
            if not os.path.exists(dst_dir):
                os.makedirs(dst_dir)
            new_folder_name = os.path.join(dst_dir, str(score)+"_"+img_name)
            original_image_file = os.path.join(dst_dir, "original")
            
            shutil.copy(matched_images_path, new_folder_name)
#             shutil.copy(p, original_image_file)
            print(p)
```
